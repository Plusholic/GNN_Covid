```python
# LSTM, RF, LGBM 넣어야함.

def each_network_plot(df, P1_pred, P2_pred, name_dict, region_dict, hop):

    num_row = len(region_dict)
    num_iter = len(P1_pred)
    fig, ax = plt.subplots(figsize=(13,13))
    P1_span_start = datetime(2021, 6, 30)
    P1_span_end = datetime(2021, 7, 19)
    P2_span_start = datetime(2021, 11, 30)
    P2_span_end = datetime(2021, 12, 19)
    
    ct = 0
    for i in range(num_row):
        ax = plt.subplot(num_row+1, 4, ct+1)
        P1_start, P1_end = datetime(2021, 6, 15), datetime(2021, 7, 19)
        df.loc[P1_start:P1_end,region_dict[i]].plot(ax=ax, color='k')
        for j in range(num_iter):
            P1_pred[j].loc[P1_start:P1_end,region_dict[i]].plot(ax=ax)
        ax.set_title(name_dict[region_dict[i]])
        ax.set_xlabel('')
        plt.setp(ax.get_xticklabels(), visible=False)
        plt.axvspan(P1_span_end, P1_span_start, facecolor='lightgray')
        ct += 1
        
        # 마지막 loop일때. i는 0부터 시작하기 때문에 마지막은 1을 빼줘야 같아짐
        if i >= num_row-2:
            ax.set_xlabel('P1', fontsize=20)
            
        ax = plt.subplot(num_row+1, 4, ct+1)
        P2_start, P2_end = datetime(2021, 11, 15),datetime(2021, 12, 19)
        df.loc[P2_start:P2_end,region_dict[i]].plot(ax=ax, color='k')
        for j in range(num_iter):
            P2_pred[j].loc[P2_start:P2_end,region_dict[i]].plot(ax=ax)
        ax.set_title(name_dict[region_dict[i]])
        ax.set_xlabel('')
        plt.setp(ax.get_xticklabels(), visible=False)
        plt.axvspan(P2_span_end, P2_span_start, facecolor='lightgray')
        ct += 1
        
        # 마지막 loop일때. i는 0부터 시작하기 때문에 마지막은 1을 빼줘야 같아짐
        if i >= num_row-2:
            ax.set_xlabel('P2', fontsize=20)
    
    plt.suptitle(f'{graph_type}', fontsize=20)
    fig.legend(['GROUND_TRUTH', 'DCRNN', 'GCN', 'STGCN'], bbox_to_anchor=(1.0, 1.0), ncol=2)
    fig.tight_layout()
    plt.savefig(f'pdf_image/{graph_type}_{hop}hop.pdf', bbox_inches='tight')
```
